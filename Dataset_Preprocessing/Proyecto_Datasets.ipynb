```python
import pandas as pd
import os
import glob
import csv
from datasets import load_dataset
```

```python
def load_file_dataframe(lang_1, lang_2, fp_lang_1, fp_lang_2):
  path = os.getcwd()

  df = pd.DataFrame(columns=["translation"])

  file_path_1 = os.path.join(path, fp_lang_1)
  file_path_2 = os.path.join(path, fp_lang_2)

  with open(file_path_1, 'r', errors='ignore') as file_1:
    with open(file_path_2, 'r', errors='ignore') as file_2:
      lines1 = file_1.read().splitlines()
      lines2 = file_2.read().splitlines()
      for i in range(len(lines1)):
        translation = {
            lang_1: lines1[i],
            lang_2: lines2[i]
        }
        new_r = {
            "translation": translation
        }
        df.loc[len(df)] = new_r

  return df
```

```python
def load_csv_dataframe(lang_1, lang_2, fp_csv):
  path = os.getcwd()

  df = pd.DataFrame(columns=["translation"])

  file_path = os.path.join(path, fp_csv)

  with open(file_path) as infile:
    reader = csv.DictReader(infile, delimiter=',')
    for row in reader:
        translation = {
            lang_1: row[lang_1],
            lang_2: row[lang_2]
        }
        new_r = {
            "translation": translation
        }
        df.loc[len(df)] = new_r

  return df
```
